```python
import imageio
import numpy as np
import matplotlib.pyplot as plt

 
```

```python
im = imageio.imread('/path/to/img/')[:,:,3]
fig,ax = plt.subplots(figsize=(10,10))
plt.imshow(im)
```

```python
XX,YY = np.meshgrid(np.arange(1200),np.arange(1200))
XX=XX.flatten()
YY=YY.flatten()
```

```python
csum = np.cumsum(im.flatten())
mx = csum[-1]
```

```python
nx=1000
```

```python
X = np.empty(shape=(nx,2))
for i in (range(nx)):
    j = np.random.randint(mx)
    k = np.searchsorted(csum,j)
    X[i]=1200-XX[k],1200-YY[k]
```

```python
fig,ax = plt.subplots(figsize=(10,10))
ax.plot(*X.T,'o')
```

```python
from annchor import BruteForce
```

```python
bf = BruteForce(X,lambda x,y:np.linalg.norm(x-y))
```

```python
bf.get_neighbor_graph()
```

```python

neighbor_graph =bf.neighbor_graph


from tqdm.notebook import tqdm as tq
fig,ax = plt.subplots(figsize=(9,9))
#ax.plot(*X.T,'.',color='k',alpha=0.25)
ax.axis('off')
k=8
for j in tq(range(k)):
    U = X[np.arange(nx)]
    V = X[neighbor_graph[0][:,j+1]]
    for u,v in zip(U,V):
        ax.plot(*np.vstack([u,v]).T,color=plt.cm.spring(j/(k-1)),zorder=-j,linewidth=2)
plt.show()
```
